```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('tips.csv')
```

```python
df.head()
```

```python
# extracting the rows with applied single condition
```

```python
# Step 1 -> Extract the column over which you want to apply condition
df['total_bill']
```

```python
# Apply condition over the column 
df['total_bill'] > 25 # return boolean value for each row value
```

```python
# Put the applied condition in the data back and rows with true boolean value will be returned
bool_value = df['total_bill'] > 25
df[bool_value]
```

```python
df[bool_value].shape
```

```python
# example 2 = customer with Female sex
```

```python
df[df['sex'] ==  'Female']
```

```python
# Providing multiple conditions at a time
# - just provide multiple condition at a time in the data
```

```python
cond1 = df['sex'] == 'Male'
cond2 = df['time'] == 'Dinner'
```

```python
# For both the conditions to be true use - &
# For one of the condition to be true - ||
```

```python
df[cond1 & cond2] # Both the conditions has to be true
```

```python
df[cond1 | cond2] # Both the conditions has to be true
```

```python
# For multiple conditions inside the same row
# Method 1 -> pass on all the condition at a time
# Method 2 -> use isin function
```

```python
df[(df['day'] == 'Sun') | (df['day'] == 'Sat') | (df['day'] == 'Thur')] # Method 1
```

```python
df[df['day'].isin(['Sat','Sun','Thur'])]
```
